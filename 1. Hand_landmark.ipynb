```python
import mediapipe as mp
import cv2
```

```python
mphands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
```

```python
Hand = mphands.Hands()
```

```python
video = cv2.VideoCapture(0)
while True:
    suc,img = video.read()
    img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result = Hand.process(img1) # to detect hand
    print(result)
    print(result.multi_hand_landmarks) # 21 landmark for each detection of hand
    # 'result' only prints the object, so inorder to get values use multi_hand_landmarks function

    # show landmark only when hand is detected
    if result.multi_hand_landmarks:
        for hand_landmark in result.multi_hand_landmarks: # since 21 landmarks, use for loop
            mp_drawing.draw_landmarks(img,hand_landmark,mphands.HAND_CONNECTIONS,mp_drawing.DrawingSpec(color=(0,0,255),
            thickness=5,circle_radius=5),mp_drawing.DrawingSpec(color=(255,0,0),thickness=5))

    cv2.imshow("image",img)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
